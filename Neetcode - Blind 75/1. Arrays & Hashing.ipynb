### Top K Elements in List
Given an integer array nums and an integer k, return the k most frequent elements within the array.

The test cases are generated such that the answer is always unique.

You may return the output in any order.

Example 1:

Input: nums = [1,2,2,3,3,3], k = 2

Output: [2,3]
Example 2:

Input: nums = [7,7], k = 1

Output: [7]
Constraints:

- 1 <= nums.length <= 10^4.
- -1000 <= nums[i] <= 1000
- 1 <= k <= number of distinct elements in nums.

```python
from typing import  List

def top_k_frequent(nums: List[int], k: int) -> List[int]: 
    nums.sort()
    count_buckets = {} 

    for num in nums: 
        if (num not in count_buckets):
            count_buckets[num] = 1 
        else:
            count_buckets[num] += 1 

    f_bucket = {} 
    ordering = [] 

    for key, val in count_buckets.items():
        if (val not in f_bucket):
            f_bucket[val] = [key]
        else:
            f_bucket[val].append(key) 
        ordering.append(val)
        ordering.sort(reverse=True)


    print(ordering, f_bucket)

    order = [] 

    for f in range(k):
        num = f_bucket[ordering[f]].pop() 
        order.append(num) 
        order.sort() 

    return order

    
print(top_k_frequent([1,1,1,2,2,2,3], 2))
print(top_k_frequent([1,2,2,3,3,3], 2))
# print(top_k_frequent([7,7], 1))
```

### Products of Array Discluding Self
Given an integer array nums, return an array output where output[i] is the product of all the elements of nums except nums[i].

Each product is guaranteed to fit in a 32-bit integer.

Follow-up: Could you solve it in 

O(n) time without using the division operation?

Example 1:

Input: nums = [1,2,4,6]

Output: [48,24,12,8]
Example 2:

Input: nums = [-1,0,1,2,3]

Output: [0,-6,0,0,0]
Constraints:

2 <= nums.length <= 1000
-20 <= nums[i] <= 20

```python
from typing import List

def product_except_self_naive(nums: List[int]) -> List[int]:
    i = j = 0 
    n = len(nums)

    products = [] 

    while (i < n):
        product_total = 1

        j = 0 

        while (j < n):
            if (j != i): 
                product_total *= nums[j]
            j = j + 1 
        
        products.append(product_total) 
        i = i + 1 
    
    return products 

def product_except_self(nums: List[int]) -> List[int]: 
    n = len(nums) 
    i = 1
    j = n - 1 - 1

    prefix = [1] * n 
    postfix = [1] * n 

    while (i < n and j >= 0):
        prefix[i] = prefix[i-1] * nums[i-1]
        postfix[j] = postfix[j + 1] * nums[j + 1]
        i += 1
        j = j - 1

    for i in range(n):
        prefix[i] = prefix[i] * postfix[i] 

    return prefix




res = product_except_self([1,2,4,6])
print(res) 
# res = product_except_self_naive([-1,0,1,2,3])
# print(res)
```

### Three Integer Sum
Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] where nums[i] + nums[j] + nums[k] == 0, and the indices i, j and k are all distinct.

The output should not contain any duplicate triplets. You may return the output and the triplets in any order.

Example 1:

Input: nums = [-1,0,1,2,-1,-4]

Output: [[-1,-1,2],[-1,0,1]]
Explanation:
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].

Example 2:

Input: nums = [0,1,1]

Output: []
Explanation: The only possible triplet does not sum up to 0.

Example 3:

Input: nums = [0,0,0]

Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.

Constraints:

- 3 <= nums.length <= 1000
- -10^5 <= nums[i] <= 10^5

Example 2:

Input: height = [2,2,2]

Output: 4
Constraints:

- 2 <= height.length <= 1000
- 0 <= height[i] <= 1000

```python
from typing import List 

def max_area(heights: List[int]) -> int: 
    maximum_area = 0 
    left = 0 
    right = len(heights) - 1 

    while (right > left): 
        width = right - left 

        if (heights[left] >= heights[right]):
            maximum_area = max(maximum_area, heights[right] * width) 
            right = right - 1 
        else:
            maximum_area = max(maximum_area, heights[left] * width)
            left = left + 1 

    return maximum_area 

# Scenario 1 - Max Height is the Left Wall 
heights_one = [5, 1, 2] 

# Scenario 2 - Max Height is the Right Wall 
heights_two = [2, 1, 5]

# Scenario 3 - Max height is the middle index 
heights_three = [2, 1, 5, 1, 2]

# res = max_area(heights_one) 
# print(res) 

# res2 = max_area(heights_two)
# print(res2)

res3 = max_area(heights_three)
print(res3) 
```
