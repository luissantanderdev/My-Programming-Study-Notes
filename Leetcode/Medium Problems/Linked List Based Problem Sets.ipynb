You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Example 1:

Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.
Example 2:

Input: l1 = [0], l2 = [0]
Output: [0]
Example 3:

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
 
Constraints:

The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.

```python
from typing import Optional, List 

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_two_numbers(l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
    sum_node = ListNode(0)

    temp1 = l1 
    temp2 = l2 
    temp3 = sum_node
    carry = 0 

    while (temp1 or temp2): 
        digit_sum = 0 

        if (temp1):
            digit_sum += temp1.val 
            temp1 = temp1.next 

        if (temp2):
            digit_sum += temp2.val
            temp2 = temp2.next

        digit_sum += carry

        if (digit_sum > 9): 
            digit_sum = digit_sum % 10
            carry = 1 
        else:
            carry = 0 
            
        temp3.next = ListNode(digit_sum) 
        temp3 = temp3.next 

    if (carry == 1):
        temp3.next = ListNode(carry) 

    return sum_node.next 


def transform_arr_to_list(nums: List) -> Optional[ListNode]: 
    node = ListNode(0)
    temp = node 
    for num in nums: 
        temp.next = ListNode(num) 
        temp = temp.next 
    return node.next 

def print_linked_list(node: Optional[ListNode]) -> List[int]:
    l = [] 
    temp = node 
    while (temp):
        print(temp.val)
        l.append(temp.val) 
        temp = temp.next 
    return l


# l1 = ListNode(2)
# l1.next = ListNode(4)
# l1.next.next = ListNode(3)

# l2 = ListNode(5)
# l2.next = ListNode(6)
# l2.next.next = ListNode(4)

# digit_sum = add_two_numbers(l1, l2)

l1 = transform_arr_to_list([9,9,9,9,9,9,9])
l2 = transform_arr_to_list([9,9,9,9])

digit_sum = add_two_numbers(l1, l2) 

arr = print_linked_list(digit_sum)
print(arr)
```
